### 5-1 extract date_range  of each profile into a list

```python
list=[]
listall_date =[0 for i in range(10000)]
for i in range(0,10000):
    c=len(experiences['jobs'][i])
    
    for j in range(0,c):
        if experiences['jobs'][i][j]['date_range'] != None:
            list.append(experiences['jobs'][i][j]['date_range'].lower())
    listall_date[i]= list.copy()
    list.clear()
listall_date[9]
```

### 5-2 Calculate the experience duration  of each profile

```python
# calculate the sum of the experiences for each profile 
# we verify first if the date range is in english or in frensh
# we calculate the date range for each experience for each profile then we sum them
# the final experience is in mounths because we found that there is some expreiences which are less than one year
fr = ["janv." ,"févr." , "mars", "avr." , "mai" , "juin" , "juil.", "août", "sept." , "oct." , "nov." , "déc." , "avr.","juil."]
filtered_fin=[]
filtredall_date =[0 for i in range(10000)]
for i in range(0,10000):
    filtered_fin.clear()
    filtered_sent.clear()
    if (listall_date[i] != []) == True :
        for j in range(0,len(listall_date[i])):
            d=experiences['jobs'][i][j]['date_range']
            if d is None:
                diff=0
            else:
                x=d.split('–')
            
                if d.find('–')==-1:
                    diff=1
                elif (x[0].strip().isdigit() == True) and ( x[1].strip().isdigit() == True):
                    diff=abs(int(x[0])-int(x[1]))*12
                else:
                    x=d.split('–')
                    s=x[0].split()
                    if x[0].strip().isdigit() == True :
                        datedeb = datetime.strptime(x[0].strip(), '%Y').date() 
                    elif (s[0]  not in fr)==True:
                        locale.setlocale(locale.LC_ALL,'en_US.UTF-8') 
                        datedeb = datetime.strptime(x[0], '%b %Y ').date()
                    else:
                        locale.setlocale(locale.LC_ALL, 'fr_FR') 
                        datedeb = datetime.strptime(x[0], '%b %Y ').date()

                
                    s1=x[1].split()
                    s1[0]
                    #print(x[1])
                    if x[1]== " Present" or x[1]== " Aujourd’hui":
                        y=date.today().strftime('%Y-%m-%d')
                        datefin= datetime.strptime(y, '%Y-%m-%d').date()
                        #print(datefin)
                    elif x[1].strip().isdigit() == True :
                        datefin = datetime.strptime(x[1].strip(), '%Y').date() 
                    elif (s1[0]  not in fr)==True:
                        locale.setlocale(locale.LC_ALL,'en_US.UTF-8') 
                        datefin = datetime.strptime(x[1], ' %b %Y').date()
                        #print(datefin)
                    else:
                        locale.setlocale(locale.LC_ALL, 'fr_FR') 
                        datefin = datetime.strptime(x[1], ' %b %Y').date()
                        #print(datefin)    
                    diff = (datefin.year - datedeb.year)*12 + abs (datefin.month - datedeb.month)
                #print(diff)
                filtered_fin.append(diff)
                        #print(filtered_fin)

    filtredall_date[i]= filtered_fin.copy()
    #print(filtredall)
    filtered_fin.clear()
        
#filtredall_date
```
